```python
#it is observed that the income is symmetrical so we can replace it with the mean
dt['Income']=dt['Income'].fillna(dt['Income'].mean())
```

```python
#check for duplicate values
dt[dt.duplicated()]
```

```python
#Convert all categorical features to numerical vectors for easy analysis
categorical_features=['Education','Marital_Status','Dt_Customer']
dt=pd.get_dummies(dt,columns=categorical_features)
```

```python
dt.head()
```

```python
#first we define our X and y variables
y=dt.Response
X=dt.drop(['Response'],axis=1)
```

```python
#it is important to scale the datasets so that the classifiers wont be bias towards values with more weights
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
# perform feature selection on the datasets
from sklearn.feature_selection import SelectKBest,f_classif
fs=SelectKBest(score_func=f_classif,k=2)
X= fs.fit_transform(X,y)
```

```python
y
```

```python
#Given we have a small dataset, it is very important to perform cross validation on the dataset
import numpy as np
from sklearn.model_selection import cross_val_score

def cv_comparison(models, X, y, cv):
    # Initiate a DataFrame for the averages and a list for all measures
    cv_accuracies = pd.DataFrame()
    maes = []
    mses = []
    r2s = []
    accs = []
    # Loop through the models, run a CV, add the average scores to the DataFrame and the scores of 
    # all CVs to the list
    for model in models:
        mae = -np.round(cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv), 4)
        maes.append(mae)
        mae_avg = round(mae.mean(), 4)
        mse = -np.round(cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv), 4)
        mses.append(mse)
        mse_avg = round(mse.mean(), 4)
        r2 = np.round(cross_val_score(model, X, y, scoring='r2', cv=cv), 4)
        r2s.append(r2)
        r2_avg = round(r2.mean(), 4)
        acc = np.round((100 - (100 * (mae * len(X))) / sum(y)), 4)
        accs.append(acc)
        acc_avg = round(acc.mean(), 4)
        cv_accuracies[str(model)] = [mae_avg, mse_avg, r2_avg, acc_avg]
    cv_accuracies.index = ['Mean Absolute Error', 'Mean Squared Error', 'R^2', 'Accuracy']
    return cv_accuracies, maes, mses, r2s, accs
```

```python
# Create the models to be tested
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)

mlr_reg = LinearRegression()
rf_reg = RandomForestRegressor(random_state=42)
xgb_reg = xgb_regressor = XGBRegressor(random_state=42)

# Put the models in a list to be used for Cross-Validation
models = [mlr_reg, rf_reg, xgb_reg]

# Run the Cross-Validation comparison with the models used in this analysis
cv_comparison(models, X_train, y_train, 4)
```

```python
#we can build our model using the RandomForestRegressor
Rfr=RandomForestRegressor(random_state=42)
Rfr.fit(X_train,y_train)
```

```python
Rfr_pred=Rfr.predict(X_test).round()
```

```python
Rfr_pred
```

```python
from sklearn import metrics
Rfr_acc=round(metrics.accuracy_score(y_test,Rfr_pred)* 100,2)
Rfr_acc
```

```python
#let us attempt for the other two classifiers
xgb=XGBRegressor(random_state=42)
xgb.fit(X_train,y_train)
xgb_pred=xgb.predict(X_test).round()
```

```python
from sklearn import metrics
xgb_acc=round(metrics.accuracy_score(y_test,xgb_pred)* 100,2)
xgb_acc
```
